### 1. (5 points)

Let $X_t = a B_{t/a^2}$, where $a \neq 0$ is constant. Show by definition that $X_t$ is also a standard Brownian motion. 

### <font color=blue> Solution 1. </font>
#### Step 1:
Since $B_t$ deontes a standard Brownian motion, $P[\omega: B_0(\omega) = 0] = 1$, i.e.$B_0 = 0$ when $t=0$. 

At the same time, when $t=0$, $t/a^2 = 0$ ($a \neq 0$ is constant). So $aB_0 = 0$, $P[\omega: aB_0(\omega) = 0] = 1$.   

Since $X_t = a B_{t/a^2}$, $P[\omega: X_0(\omega) = 0] = 1$, the process starts at zero almost surely.
#### Step 2:
For any $0 \leq s < t$ and $m<n$, $$
P[m \leq B_t - B_s \leq n] = \frac{1}{\sqrt{2\pi(t-s)}}\int_m^n e^{-\frac{x^2}{2(t-s)}}dx.
$$
Let $t=t/a^2$ and $s=s/a^2$, we can get $$
P[m \leq B_{t/a^2} - B_{s/a^2} \leq n] = \frac{1}{\sqrt{2\pi(t/a^2-s/a^2)}}\int_m^n e^{-\frac{x^2}{2(t/a^2-s/a^2)}}dx = \frac{1}{\sqrt{2\pi(t-s)}}\int_m^n e^{-\frac{(ax)^2}{2(t-s)}}d(ax).
$$
So the random variable $B_{t/a^2} - B_{s/a^2}$  is normally distributed with mean 0 and variance $t-s$. 

Since the increments $B_{t/a^2} - B_{s/a^2}$ are Gaussian (standard Brownian motion property), multiplying the constant $a$ does not affect their Gaussian nature.   

Therefore, $X_t - X_s$ is also normally distributed with mean 0 and variance $t-s$.
#### Step 3:
Since $B_t$ has independent increment, i.e., for any $0\leq t_1 < t_2 < \cdots < t_n$, the random variables

  $$
    B_{t_1}, \; B_{t_2} - B_{t_1},\; \cdots, \; B_{t_n} - B_{t_{n-1}}
  $$  
are independent.  

Let $t=t/a^2$, for any $0\leq {t/a^2}_1 < {t/a^2}_2 < \cdots < {t/a^2}_n$, the random variables

  $$
    B_{{t/a^2}_1}, \; B_{{t/a^2}_2} - B_{{t/a^2}_1},\; \cdots, \; B_{{t/a^2}_n} - B_{{t/a^2}_{n-1}}
  $$
are independent.  

As multiplying by a constant $a$ does not affect the independence, $X_t = a B_{t/a^2}$ has independent increment too.
#### Step 4:

$X_{t + \Delta t} - X_t = aB_{(t + \Delta t)/a^2} - aB_{t/a^2}$

As $\Delta t \to 0$, $(t + \Delta t)/a^2 \to t/a^2$,

$\lim_{{\Delta t}\to 0} (X_{t + \Delta t} - X_t) = \lim_{{\Delta t}\to 0} aB_{(t + \Delta t)/a^2} - aB_{t/a^2} = aB_{t/a^2} - aB_{t/a^2} = 0$.

Since the limit of the increment is zero as $\Delta t$ approaches zero, $X_t$ is continuous.


#### Above of all, $X_t$ is also a standard Brownian motion. 

## Donsker's invariance principle

Recall the Donsker's invariance principle: 


Suppose $\{X_i\}_{i=1}^\infty$ is an iid sequence of random variables with mean 0 and and variance 1. Let $S_n = \sum_{i=1}^n X_i$. Define the function $\mathfrak{S}_n$ of $t$ by 

$$
  \mathfrak{S}_n(t) = \frac1{\sqrt n}\sum_{i=1}^n\left[S_{i-1} + n \left( t-\frac{i-1}n \right) X_i \right]\,\1_{\left(\frac{i-1}n,\frac in\right]}(t) .
$$

In fact, $\mathfrak{S}_n$ is simply the linear interpolation of the scaled random walk $\left\{\frac{S_1}{\sqrt n}, \frac{S_2}{\sqrt n}, \cdots, \frac{S_n}{\sqrt n} \right\}$.

Then, $\mathfrak{S}_n\Longrightarrow W$ as $n\to\infty$, where $W$ denotes a standard Brownian motion. 

In other words, as $n\to\infty$, the linearly interpolated scaled random walk $\mathfrak{S}_n$ converges weakly to a standard Brownian motion.

In this problem, we show, numerically by simulation, the Donsker's invariance principle with the iid sequence $X_i$ being 

- (a) $X_i \sim N(0, 1)$, i.e., the standard normal distribution
- (b) $X_i \sim \chi^2(1.5)$, i.e., chi-squared distribution with degress of freedom 1.5.

Consider the Brownian motion in the time interval $[0,1]$, use $50$ time steps and generate $20,000$ sample paths to reproduce the plots for sample mean, sample standard deviation, and the histogram as given in the lecture. To begin with, please set the random seed by including ``numpy.random.seed(3141)`` on top of your code. 

```python
# as always, import required modules and functions
import numpy as np
from numpy import sqrt, log, exp
import matplotlib.pyplot as plt
import scipy.stats as ss
from scipy.stats import norm
import seaborn as sns
```

### <font color=blue> Solution 2(a). </font>

```python
# insert python code for (2a) in this cell
np.random.seed(3141)

# number of samples, number of steps, terminal time
n_sim, n_steps, T = 20_000, 50, 1  
dt = T/n_steps

# initialize the Brownians
B = np.zeros([n_sim, n_steps + 1])

# simulation step
for i in range(n_steps):
    dB = norm.rvs(size=n_sim)
    dB = dB - dB.mean()  # now dB has mean 0
    dB = dB / dB.std()  # now dB has variance 1
    dB = np.sqrt(dt) * dB
    B[:, i + 1] = B[:, i] + dB

# plot
t = np.arange(0, T + dt, dt)
# t = np.linspace(0, T, n_steps+1)
plt.figure(figsize=(10, 8))


# histograms
plt.subplot(2, 2, 3)
time = 0.2
sns.histplot(B[:, t == time], bins=50, stat='density', label=f'$t={time}$')
sns.histplot(B[:, -1], bins=50, stat='density', color='orange', label=f'$t={T}$')
x = np.linspace(B[:, -1].min(), B[:, -1].max(), 100)
plt.plot(x, norm.pdf(x), 'r--', lw=1, label='normal pdf')
plt.plot(x, norm.pdf(x, scale=sqrt(time)), 'y--', lw=1, label='normal pdf')
plt.xlabel('$B$')
plt.legend();

# sample mean
plt.subplot(2, 2, 1)
plt.plot(t, B.mean(axis=0))
plt.hlines(y=0, xmin=0, xmax=T, color='red', ls='dotted')
plt.xlabel('$t$')

# sampel standard deviation
plt.subplot(2, 2, 2)
plt.plot(t, B.std(axis=0), label='sample std')
plt.plot(t, np.sqrt(t), 'r--', label='sqrt t')
plt.xlabel('$t$')
plt.legend();

```
